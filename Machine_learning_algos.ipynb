```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
###converting into labels
labels = LabelEncoder().fit(df_train['species'])
y_train = np.array(labels.transform(df_train['species']))
```

```python
x_train =np.array(df_train.drop(['id','species'],axis=1))
```

```python
##Normalization of the features
x_train_norm = (x_train-np.amin(x_train,0))/(np.amax(x_train,0)-np.amin(x_train,0))
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x_train_norm, y_train,test_size=0.20,random_state=0)
```

```python
''' 
As a first insight we apply almost all the classifiers and see the test results
'''
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
classifiers = [LinearDiscriminantAnalysis(),
               LogisticRegression(),
               KNeighborsClassifier(3),
               SVC(kernel="linear"),
               AdaBoostClassifier(),
               GradientBoostingClassifier(),
               RandomForestClassifier(),
               DecisionTreeClassifier(),]
```

```python
for clf in classifiers:
    #Training classifier
    print(clf)
    clf.fit(train_x,train_y)
    print("____")
    #testin classifier on test data
    predictions = clf.predict(test_x)
    accuracy = accuracy_score(test_y, predictions) * 100
    print("validation Accuracy: " + str(accuracy) + "%")
    print("____")
    print("____")
    print("____")
    print("____")
```
